## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
import pandas as pd
import json
import numpy as np
from pandas.io.json import json_normalize
df = pd.read_json('data/world_bank_projects.json')
df['mjtheme_namecode'].groupby['code', 'name'].size()
```

```python
df.loc[df.code == '1', 'name'] = 'Economic Management'
df.loc[df.code == '2', 'name'] = 'Public sector governance'
df.loc[df.code == '3', 'name'] = 'Rule of law'
df.loc[df.code == '4', 'name'] = 'Financial and private sector development'
df.loc[df.code == '5', 'name'] = 'Trade and integration'
df.loc[df.code == '6', 'name'] = 'Social protection and risk management'
df.loc[df.code == '7', 'name'] = 'Social dev/gender/inclusion'
df.loc[df.code == '8', 'name'] = 'Human development'
df.loc[df.code == '9', 'name'] = 'Urban development'
df.loc[df.code == '10', 'name'] = 'Rural development'
df.loc[df.code == '11', 'name'] = 'Environment and natural resources management'
```

```python
df['mjtheme_namecode'].head(15)
```

```python
global_dict = dict()
for x in df['mjtheme_namecode']:
    for y in x:
        if y["name"]!="":
            global_dict[y["code"]] = y["name"]
            
```

```python
def fill_missing_name(x):
    temp_list = list()
    for y in x:
        temp_dict = dict()
        temp_dict["code"] = y["code"]
        temp_dict["name"] = global_dict[y["code"]]
        temp_list.append(temp_dict)
    return temp_list

df['mjtheme_namecode'] = df['mjtheme_namecode'].map(fill_missing_name)
```

```python
from collections import Counter
def predominant_theme(data):
    theme = Counter(k['mjtheme_namecode'] for k in data if k.get('mjtheme_namecode'))
    for theme, count in theme.most_common():
        print(theme, count)
```
